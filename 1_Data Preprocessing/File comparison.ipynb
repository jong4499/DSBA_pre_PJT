```python
import os
import cv2
from glob import glob
import pandas as pd
import numpy as np
```

```python
img_lst = glob('C:/Users/admin/Desktop/tester/*.jpg')
label_lst = glob('C:/Users/admin/Desktop/tester/*.txt')
print(len(img_lst))
print(len(label_lst))
```

```python
comp_img = img_lst.copy()
comp_label = label_lst.copy()
print(len(comp_img))
print(len(comp_label))
```

```python
new_comp_img=[]
for i in range(len(comp_img)):
  new_comp_img.append(comp_img[i].split('.'))

new_comp_label=[]
for i in range(len(comp_label)):
  new_comp_label.append(comp_label[i].split('.'))


print(new_comp_img[10])
print(new_comp_label[10])
```

```python
df_comp_img = pd.DataFrame(new_comp_img)
df_comp_label = pd.DataFrame(new_comp_label)

df_comp_img.columns = ['name', 'exten']
df_comp_label.columns = ['name', 'exten']
```

```python
def GetDifferences(df1, df2):
  df = pd.concat([df1, df2]).reset_index(drop=True)
  idx = [diff[0] for diff in df.groupby(list(df.columns)).groups.values() if len(diff) == 1]
  return df.reindex(idx)
```

```python
answer = GetDifferences(df_comp_label, df_comp_img)
print(len(answer))
```
